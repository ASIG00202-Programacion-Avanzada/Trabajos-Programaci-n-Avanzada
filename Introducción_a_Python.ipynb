## Bucles: for, while y list comprehensions

```python
l = range(5)
```

```python
l
```

```python
type(l)
```

```python
l = list(l)
```

```python
l
```

```python
type(l)
```

For loops

```python
for i in l:
    print(f'the number is: {i}')
```

Break

```python
for i in l:
    print(f'the number is: {i}')
    if i >= 3:
        print("breaking the loop")
        break
```

Continue

```python
for i in l:
    if i == 1:
        continue
    print(f"{i}")
```

```python
for i, j in enumerate(['a', 'b']):
    print(f"index: {i} ; value {j}")
```

Else en for loops

```python
for i in range(5):
    if i > 5:
        break
else:
    print("Didn't break")
```

While loops

```python
i = 0
while i < 5:
    i += 1
    print(i)
```

```python
i = 1
while i > 0:
    i += 1
    print(i)
    if i == 10:
        break
```

List y Dict Comprehension

```python
lc = [i.upper() for i in ('a', 'b')]
print(lc)
```

```python
[i for i in ('a', 'b') if i != 'b']
```

```python
dc = {k.lower(): v + 1 for k,v in {'A': 1, 'B': 2}.items()}
dc
```

## Clases y Objetos

```python
class Human:
    
    """Friendly human with a name."""

    def __init__(self, name):
        self.name = name
    
    def greet(self):
        
        print(f"Hi! My name is {self.name}")
```

```python
# instanciamos
jose = Human('jose')
```

```python
# hacemos una llamada a un metodo
jose.greet()
```

```python
# podemos acceder a un atributo de manera directa
jose.name
```

```python
# podemos crear nuevos atributos de manera dinámica sin reestricciones
jose.last_name = 'perez'
```

```python
jose.last_name
```

```python
# utilizamos la función sum que es una función built in del lenguaje Python
sum
```

```python
# de esta forma podemos agregarle un nuevo método al objeto en tiempo de ejecución 
jose.suma = sum
```

```python
# este nuevo método lo podemos utilizar como cualquier otro
jose.suma([1, 2])
```
